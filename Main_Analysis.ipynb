```python
from zipfile import ZipFile
import numpy as np
import pandas as pd
import zipfile
import re
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure



zf = zipfile.ZipFile('Data_incidents.csv.zip') 
data = pd.read_csv('Data_incidents.csv.zip').reset_index()
```

```python
data = data.drop(columns=['index','NAV','Unnamed: 0']) # drop unwanted columns
```

```python
data['Date'] = pd.to_datetime(data.date.str.strip(), format = '%Y-%m-%d') # make date_time
data = data.drop(['date'], axis=1);
```

```python
data['Year'] = data.Date.apply(lambda x: x.year) # make Year column
```

```python
data = data[data['Year']!=2013] # drop 2013 from observations
```
