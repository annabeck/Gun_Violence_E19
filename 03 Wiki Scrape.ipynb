```python
import requests
from bs4 import BeautifulSoup
import os
import codecs
import lxml.html as lh
import pandas as pd
```

```python
url = requests.get('https://en.wikipedia.org/wiki/Gun_laws_in_the_United_States_by_state')
doc = lh.fromstring(url.content)
tr_elements = doc.xpath('//tr')
len(tr_elements)
```

```python
#remove irrelevant tables
tr = []

for T in tr_elements:
    if len(T) ==5:
        tr.append(T)
       # tr_elements.remove(T)
```

```python
[len(T) for T in tr[:12]]
```

```python
#Create empty list
col=[]
i =0
#For each row, store each first element (header) and an empty list

for t in tr[0]:
    i+=1
    name=t.text_content()
            #print '%d:"%s"'%(i,name)
    col.append((name,[]))
```

```python
col
```

```python
for t in tr:
    for i in range(5):
        test = t[i].text_content()
    #print(test)
        col[i][1].append(test)
```

```python
[len(C) for (title,C) in col]
```

```python
Dict={title:column for (title,column) in col}
df=pd.DataFrame(Dict)
```

```python
df.head(20)
df.to_csv('Data/state_gunlaw.csv', index=True, sep = ',')
```
