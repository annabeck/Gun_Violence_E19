```python
import requests
from json2table import convert
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = 'https://api.census.gov/data/'
key = '3b73c25c2be006e16d34baf9a203112b51cacdbf'
```

```python
year = [2014,2015,2016,2017,2018]
dataset = '/acs/acs1'
var = 'B01001_001E,B01001_002E,B01001_026E,B01002_001E,B02001_002E,B02001_003E,B02001_004E,B02001_005E,B02001_006E,B02001_008E,B19055_002E,B19083_001E,B19301_001E'
geo = 'state'
geo_1 = ':*'
```

```python
col = [('Year', []),
 ('NAME', []),      
 ('Total_Pop', []),
 ('Male_Pop', []),
 ('Female_Pop', []),
 ('Age_Median', []),
 ('White_Pop', []),
 ('Black_AA_Pop', []),
 ('AmericanIndian_AlasNative_Pop', []),
 ('Asian_Pop', []),
 ('PacificNative_Pop', []),
 ('OtherRace_Pop', []),
 ('Social_Security', []),
 ('Gini', []),
 ('PerCapita_Income', []),
 ('state', [])]
 #('county', [])]

for y in year:
    
    full_url = url+str(y)+dataset+'?&get=NAME,'+ var + '&for=' +geo+geo_1
    print(full_url)
    responce = requests.get(full_url).json()
    
    for t in responce:
        col[0][1].append(y)
        for i in range(len(responce[0])):
            val = t[i]
            col[i+1][1].append(val)
            
    
    
```

```python
Dict={title:column for (title,column) in col}
df=pd.DataFrame(Dict)
data = df[df['NAME']!='NAME']
```

```python
data
```

```python
data = data.drop(columns=['Age_Median', 'Gini'])
```

```python
data.iloc[:,2:-1] = data.iloc[:,2:-1].astype(int)
```

```python
#data['Median_Age'] = pd.to_numeric(data.Age_Median, downcast='float')
#data['Gini_coef'] = pd.to_numeric(data.Gini, downcast='float')
#data['population'] = pd.to_numeric(data.Total_Pop, downcast='float')

#data = data.drop(columns=['Age_Median','Gini','Total_Pop'])
data = data[data['NAME']!='Alaska']
data = data[data['NAME']!='Hawaii']
data = data[data['NAME']!='District of Columbia']
```

```python
data_2018 = data[data['Year'] == 2018].reset_index()
#state_demo = data.groupby('NAME').mean().reset_index()
```

```python
fig, ax = plt.subplots(figsize=(12,8))
data_2018.plot.bar(x='NAME', y='Total_Pop', rot=0, ax=ax)
plt.xticks(x='NAME', rotation='vertical');
```

```python
data_2018['white_share'] = data_2018.White_Pop / data_2018.Total_Pop
data_2018['aa_share'] = data_2018.Black_AA_Pop / data_2018.Total_Pop
data_2018['ai_an_share'] = data_2018.AmericanIndian_AlasNative_Pop / data_2018.Total_Pop
data_2018['asian_share'] = data_2018.Asian_Pop / data_2018.Total_Pop
data_2018['pn_share'] = data_2018.PacificNative_Pop / data_2018.Total_Pop
data_2018['other_share'] = data_2018.OtherRace_Pop / data_2018.Total_Pop
```

```python
fig, ax = plt.subplots(figsize=(15,8))

p1 = plt.bar(data_2018.NAME, data_2018.white_share)
p2 = plt.bar(data_2018.NAME, data_2018.aa_share, bottom=data_2018.white_share, edgecolor = 'white', linewidth = '1')
p3 = plt.bar(data_2018.NAME, data_2018.ai_an_share, bottom=data_2018.white_share+data_2018.aa_share, edgecolor = 'white', linewidth = '1')
p4 = plt.bar(data_2018.NAME, data_2018.asian_share, bottom=data_2018.white_share+data_2018.aa_share+data_2018.ai_an_share, edgecolor = 'white', linewidth = '1')
p5 = plt.bar(data_2018.NAME, data_2018.pn_share, bottom=data_2018.white_share+data_2018.aa_share+data_2018.ai_an_share+data_2018.asian_share, edgecolor = 'white', linewidth = '1')
p6 = plt.bar(data_2018.NAME, data_2018.other_share, bottom=data_2018.white_share+data_2018.aa_share+data_2018.ai_an_share+data_2018.asian_share+data_2018.pn_share, edgecolor = 'white', linewidth = '1')

plt.xticks(x='NAME', rotation='vertical');
plt.legend((p6[0],p5[0],p4[0], p3[0], p2[0], p1[0]), ('Other', 'Pacific', 'Asian','Indian/Alaskan','African-American', 'White'),prop={'size': 15})
```
